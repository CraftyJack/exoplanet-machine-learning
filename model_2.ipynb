```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.

# column definitions can be found here: https://exoplanetarchive.ipac.caltech.edu/docs/API_kepcandidate_columns.html
# we'll exclude all the second error terms, as well as the disposition
column_names = list(df.columns.values)

exclusion_list = ['koi_disposition', '_err2',]
selected_columns = [x for x in column_names if all(y not in x for y in exclusion_list)]

selected_features = df[selected_columns]
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
X = selected_features
y = df.koi_disposition

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
X_train.shape, y_train.shape
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import LabelEncoder, StandardScaler
X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Create LogisticRegression Model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter = 1000)
```

# Train the Model



```python
classifier.fit(X_train_scaled, y_train)
coef = classifier.coef_[0]
classifier
```

```python
col_names = X.columns
summary_df = pd.DataFrame({'feature_name': col_names, 'correl': coef})
summary_df.sort_values(['correl'], ascending=False)
```

```python
print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
import numpy as np

C = [1, 5, 10, 50, 100]

param_grid = dict(C=C)

grid = GridSearchCV(classifier, param_grid, cv=5, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
predictions = grid.predict(X_test_scaled)
print('Test Acc: %.3f' % grid.score(X_test_scaled, y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=y_test.unique()))
```
